```python
import pandas as pd
df = pd.read_csv('prediction_credit.csv', index_col=False, header=0)
df.head(10)
```

```python
nb_inf = 0
nb_interation = 0
serie = df.transpose() # here we convert the DataFrame into a serie
for ligne in serie: # For each row in the df
    if serie[ligne][11] == 1: # If the value in column 11 SeriousDlqin2yrs = 1 
        proba_positif = serie[ligne][19] # We stocke 1 probability value into proba_positif
        for ligne_neg in serie: # For negative value
            if serie[ligne_neg][11] == 0: # If the value in column 11 SeriousDlqin2yrs = 0
                if proba_positif > serie[ligne_neg][19]: # If proba_positif value is bigger than 1 probability of ligne_neg
                    nb_inf += 1 # we add each round 1
                nb_interation += 1 # We add the number of loop (denominator of AUC formula)
                    
print(nb_inf)
print(nb_interation)
print(f"{nb_inf/nb_interation}")
# nb_inf / nombre de ligne de la serie = AUC
```

```python
pos = df.loc[df['SeriousDlqin2yrs'] == 1] # loc. to localize values 1 in the 'SeriousDlqin2yr' column
neg = df.loc[df['SeriousDlqin2yrs'] == 0] # loc. to localize values 0 in the 'SeriousDlqin2yr' column
x = 0
y = 0

# for every positive values in the  positive 1 proba column 
for threshold_pos in pos['1 probability']:
# for every neg values in the  negative 1 proba column 
    for threshold_neg in neg['1 probability']:
        if threshold_pos > threshold_neg:
            x += 1 
        y += 1 
print(x/y) Auc formula
```
